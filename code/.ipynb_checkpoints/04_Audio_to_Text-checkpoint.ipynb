## Convert Audio to Text

```python
# instantiate dataframe
audio_to_text_df = pd.DataFrame(columns=['id','audio_to_text'])
audio_to_text_df
```

```python
deleted_vids = [62,86,231,373,413] # videos that were deleted btw scraping url and video info
```

```python
for index in range (0,len(df)):
    if index not in deleted_vids: # only run if the video is not a deleted vid
        speech = {}
        speech['id'] = str(index) + '_post' 

        try:
            print(f'converting video number: {index}')
            # convert mp4 to wav
            mp4_to_wav_cmd = f'ffmpeg -i ../videos/{index}.mp4 ../videos/{index}.wav'

            # execution video conversion command using os
            os.system(mp4_to_wav_cmd)

            if len(AudioSegment.from_wav(f'../videos/{index}.wav')) < 56000: # video <56k millisecond
            
                r = sr.Recognizer()
                with sr.AudioFile(f'../videos/{index}.wav') as source:
                     audio = r.record(source, duration=220) 
                speech['audio_to_text'] = r.recognize_google(audio, language = 'en-US')
                # save info to audio_to_text_df
                audio_to_text_df.loc[len(audio_to_text_df)] = speech
                
            else: # video >= 56k millisecond
                # break into different segments since speech_recognition only accepts <10mb
                sound = AudioSegment.from_wav(f'../videos/{index}.wav')
                # no. of segments to break into
                no_of_segments = ceil(len(sound) / 56000)
                join_segment = ''
                for segment in range(no_of_segments):
                    split_sound = sound[(segment * 56000):((segment + 1) * 56000)]
                    split_sound.export(out_f = '../videos/temp.wav',format = 'wav')
                    r = sr.Recognizer()
                    with sr.AudioFile('../videos/temp.wav') as source:
                        audio = r.record(source, duration=220)
                    segment = r.recognize_google(audio, language = 'en-US')
                    join_segment += segment
                # save info to audio_to_text_df
                speech['audio_to_text'] = join_segment
                audio_to_text_df.loc[len(audio_to_text_df)] = speech
                
        except sr.UnknownValueError:
            speech['audio_to_text'] = ''
            # save info to audio_to_text_df
            audio_to_text_df.loc[len(audio_to_text_df)] = speech
            pass
```
